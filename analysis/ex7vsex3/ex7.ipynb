```python
import os
import json
from pathlib import Path
import matplotlib.pyplot as plt

# ====== Ë®≠ÂÆö ======
BASE_DIR = Path("/home/toshiya1048/dev/QA_knap/out/ex7")   # ex7„Éï„Ç©„É´„ÉÄ
PLOT_DIR = BASE_DIR / "boxplots"
PLOT_DIR.mkdir(parents=True, exist_ok=True)

def load_json_safe(p: Path):
    try:
        with open(p, "r") as f:
            return json.load(f)
    except Exception as e:
        print(f"[warn] failed to load {p}: {e}")
        return None

def iter_iteration_files(base: Path):
    for ts_dir in sorted([d for d in base.iterdir() if d.is_dir()]):
        for inst_dir in sorted([d for d in ts_dir.iterdir() if d.is_dir() and d.name.endswith('_before_data')]):
            instance = inst_dir.name.replace('_before_data', '')
            for f in sorted(inst_dir.glob('iteration_*.json')):
                if 'timings' in f.name:
                    continue
                try:
                    it = int(f.stem.split('_')[1])
                except Exception:
                    continue
                yield instance, it, f

def collect_distributions(base: Path):
    out = {}
    for instance, it, fp in iter_iteration_files(base):
        data = load_json_safe(fp)
        if not isinstance(data, list):
            continue
        clusters, dists = [], []
        for rec in data:
            cid = rec.get("cluster_id")
            vals = []
            runs = rec.get("runs") or []
            for r in runs:
                if isinstance(r, dict) and "total_distances" in r:
                    try:
                        vals.append(float(r["total_distances"]))
                    except Exception:
                        pass
            if not vals:
                td = rec.get("total_distance")
                if isinstance(td, (int, float)):
                    vals = [float(td)]
            if cid is not None and vals:
                clusters.append(cid)
                dists.append(vals)
        if clusters and dists:
            order = sorted(range(len(clusters)), key=lambda i: clusters[i])
            clusters = [clusters[i] for i in order]
            dists = [dists[i] for i in order]
            out[(instance, it)] = {"clusters": clusters, "dists": dists, "src": fp}
    return out

def plot_box_for(instance, iteration, clusters, dists, save_dir):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.boxplot(dists, positions=range(1, len(clusters)+1), showfliers=True)
    ax.set_title(f"{instance} ‚Äî Iteration {iteration} (per-cluster route length)")
    ax.set_xlabel("Cluster ID")
    ax.set_ylabel("Total Distance (per run)")
    ax.set_xticks(range(1, len(clusters)+1))
    ax.set_xticklabels([str(c) for c in clusters])
    ax.grid(True)
    plt.tight_layout()

    # ‰øùÂ≠òÔºãNotebookË°®Á§∫
    out_path = save_dir / f"box_{instance}__iter{iteration}.png"
    plt.savefig(out_path, dpi=200)
    plt.show()  # ‚Üê Jupyter „ÅßÁõ¥Êé•Ë°®Á§∫
    return out_path

# ===== ÂÆüË°å =====
results = collect_distributions(BASE_DIR)
saved = []

for (instance, it), payload in sorted(results.items(), key=lambda x: (x[0][0], x[0][1])):
    out_png = plot_box_for(instance, it, payload["clusters"], payload["dists"], PLOT_DIR)
    saved.append(out_png)

print(f"‚úÖ Âá∫ÂäõÂÆå‰∫Ü ({len(saved)} plots)\n‰øùÂ≠òÂÖà: {PLOT_DIR}")
```
